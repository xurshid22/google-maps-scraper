```
# Step 3: Set your search query
QUERY = 'restaurants Moscow'   # <-- Change this!
MAX_RESULTS = 20               # <-- Change this!

print(f'Query: {QUERY}')
print(f'Max results: {MAX_RESULTS}')
```

```
# Step 4: Run scraper
import sys
sys.argv = ['scraper', QUERY]

# Patch max_results
import google_maps_scraper as gms
results = gms.scrape_google_maps(QUERY, MAX_RESULTS)
print(f'\n✅ Found {len(results)} businesses!')
```

```
# Step 5: Save and download results
import pandas as pd
from google.colab import files

if results:
    df = pd.DataFrame(results)
    filename = f'results_{QUERY.replace(" ", "_")}.csv'
    df.to_csv(filename, index=False, encoding='utf-8-sig')
    print(df[['name','website','phone','rating']].to_string())
    print(f'\n💾 Downloading {filename}...')
    files.download(filename)
else:
    print('No results found')
```
